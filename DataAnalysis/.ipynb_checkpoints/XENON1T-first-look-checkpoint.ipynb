## Access to data
The XAMS (and XENON1T) data are stored in collections of zip-files, stored in directories. The module Zip knows how to read these zip files and provides you a few methods to access the Events

```python
import Zip
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import time

raw = Zip.ReadZipped("/data/nikhefproject/data/xenon1t/NG") # Neutron Generator run
```

### Reading a Single Event
The next cell shows you how to read in a single Event. The Event is a structure that contains all the information that was obtained in a short amount of time (<200 us). Have a look at the pax/datastructure.py file for more details. Below are some of the event properties read in from the data file.

```python
event = raw.get_single_event(1)
print(event)
```

You can access these elements from the Event. You can for instance print the time of the event in nanoseconds since January 1, 1970: 

```python
print("Event start:", event.start_time, " stop: ", event.stop_time)
print("First event at:", time.asctime(time.gmtime(int(event.start_time/1e9))), " UTC")
```

Most importantly, it also gives you access to the data. The waveforms (similar to Oscilloscope traces) are stored in Pulse objects that are specific to a channel and time segment. Have a look at pax/datastructure.py for all the details. Here is the first pulse in the list

```python
pulse = event.pulses[0]
print(pulse)
```

### Pulses
pulse.left gives you the index/sample of the left bound and pulse.right of the right bound. The array gives the ADC value for each sample, starting for the index position of pulse.left and going all the way to pulse.right. The ADC samples make up the **waveform** and provide the "Oscilloscope trace" of the PMT signal. Note: for XENON1T, the baseline (i.e. no signal excursion) **fluctuates around 16000** (different from XAMS!), where the fluctuations are caused by (electronic noise). A signal will fluctuate downward (make the ADC values smaller). Keep in mind that Dark Matter physicists often show these same signals as *positive* fluctuations (e.g. Erik's thesis). For that you subtract the baseline and invert the signal.  

```python
%matplotlib inline
t = np.linspace(pulse.left, pulse.right, pulse.right-pulse.left+1)
figs, ax = plt.subplots()
ax.plot(t, pulse.raw_data)
ax.set_xlabel("time [sample]")
ax.set_ylabel("ADC")
```

These fluctuations are noise. Here is what a real signal looks like:  

```python
pulse = event.pulses[18] # I knew where to look in this event
t = np.linspace(pulse.left, pulse.right, pulse.right-pulse.left+1)
figs, ax = plt.subplots()
ax.plot(t, pulse.raw_data)
ax.set_xlabel("time [sample]")
ax.set_ylabel("ADC")
```

### S1 and S2 signals
The above signal is an example of an **S1** signal coming from direct xenon scintillation light. This event also has an **S2** signal. These are the signals that come from the electrons that were released in the interaction with the xenon nucleus, then drifted through the LXe volume due to the electric drift-field and then extracted from the LXe liquid and further amplified in the high E-field in the xenon gas above the liquid. This causes additional scintillation and the signals are _much_ higher (it's event clipped!):   

```python
pulse = event.pulses[24] # I knew where to look in this event
t = np.linspace(pulse.left, pulse.right, pulse.right-pulse.left+1)
figs, ax = plt.subplots()
ax.plot(t, pulse.raw_data)
ax.set_xlabel("time [sample]")
ax.set_ylabel("ADC")
```

# Summed Waveform
The Summed Waveform in Channel 254 is special. It is essentially an analog addition of all the channels. You could for instance use it to identify the global S1 and S2 and use it to then go into the individual channels to do a more accurate pulse shape analysis. 

```python
event = raw.get_single_event(4)

n = 200000 # ample size for full waveform 
ts = np.ones(n, dtype=np.int16)
ts = ts * 16382 # more or less baseline for this ADC
t = np.linspace(0, n, n) # 10ns sampling ADC

for p in event.pulses:
    ch = p.channel
    if ch == 254:
        print(ch)
        i = 0
        for adc in p.raw_data:
            ts[p.left + i] = adc
            i = i + 1
figs, ax = plt.subplots()
figs.set_size_inches(10,10)
ax.plot(t, ts, label='Sum Waveform')
ax.set_xlim(90000,120000)
```
